```python
import os
import pandas as pd
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 1000)
```

```python
path = os.getcwd()+"/monthly"
```

```python
from os import walk
dirs = []
for (dirpath, dirnames, filenames) in walk(path):
    dirs = dirnames
    break
```
